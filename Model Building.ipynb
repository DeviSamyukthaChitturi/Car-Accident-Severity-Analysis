## Data Visualization

#### Visualization - 1: Correlation matrix between different variables related to accidents, casualties, and vehicles dataframes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Merge Accident, Casualties, and Vehicles dataframes on 'Accident_Index'
merged_df = pd.merge(Accidents, Casualties[['Accident_Index', 'Age_of_Casualty', 'Casualty_Severity']], on='Accident_Index')
merged_df = pd.merge(merged_df, Vehicles[['Accident_Index', 'Vehicle_Type', 'Skidding_and_Overturning', 'Sex_of_Driver', 'Age_of_Driver']], on='Accident_Index')

# Select relevant columns for correlation analysis
correlation_df = merged_df[['Accident_Severity', 'Number_of_Vehicles', 'Number_of_Casualties', 'Road_Type',
                            'Speed_limit', 'Light_Conditions', 'Weather_Conditions', 'Road_Surface_Conditions',
                            'Urban_or_Rural_Area', 'Did_Police_Officer_Attend_Scene_of_Accident',
                            'Age_of_Casualty', 'Casualty_Severity', 'Vehicle_Type', 'Skidding_and_Overturning',
                            'Sex_of_Driver', 'Age_of_Driver']]

# Compute the correlation matrix
correlation_matrix = correlation_df.corr()

# Create a heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Accident, Casualties, and Vehicles Data', fontsize = 16)
plt.show()
```

#### Visualization - 2: To visualize the Count of Accidents based on the Accident Severity

```python
# Merge Accident and Casualties dataframes on 'Accident_Index'
merged_df = pd.merge(Accidents, Casualties[['Accident_Index', 'Age_of_Casualty', 'Casualty_Severity']], on='Accident_Index')

# Count the occurrences of each accident severity level
severity_counts = merged_df['Accident_Severity'].value_counts()

# Define custom colors for each severity level
custom_colors = ['lightgreen', 'salmon', 'skyblue']

# Create a pie chart to visualize the distribution of accident severity levels
plt.figure(figsize=(8, 8))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%', colors=custom_colors)
plt.title('Distribution of Accident Severity Levels', fontsize=16)

# Add a legend to indicate the colors
plt.legend(title='Severity Level', loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

#### Visualization - 3: To visualize the Accident Rates based on different Causes for the Accidents

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selected columns from the Accident dataframe
selected_columns = ['Road_Type', 'Speed_limit', 'Light_Conditions', 'Weather_Conditions', 'Road_Surface_Conditions', 'Urban_or_Rural_Area']

# Corresponding labels for each unique value in the selected columns
labels = {
    'Road_Type': {1: 'Roundabout', 2: 'One way street', 3: 'Dual carriageway', 6: 'Single carriageway', 7: 'Slip road', 9: 'Unknown'},
    'Light_Conditions': {1: 'Daylight', 4: 'Darkness - lights lit', 7: 'Darkness - lighting unknown', 5: 'Darkness - lights unlit', 6: 'Darkness - no lighting'},
    'Weather_Conditions': {1: 'Fine no high winds', 2: 'Raining no high winds', 3: 'Snowing no high winds', 4: 'Fine + high winds', 5: 'Raining + high winds',
                            6: 'Snowing + high winds', 7: 'Fog or mist', 8: 'Other', 9: 'Unknown'},
    'Road_Surface_Conditions': {1: 'Dry', 2: 'Wet or damp', 3: 'Snow', 4: 'Frost or ice', 5: 'Flood over 3 cm. deep'},
    'Urban_or_Rural_Area': {1: 'Urban', 2: 'Rural'}
}

# Create subplots for each column
fig, axes = plt.subplots(3, 2, figsize=(16, 12))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot count plots for each selected column
for i, column in enumerate(selected_columns):
    sns.countplot(x=column, data=Accidents, ax=axes[i], palette='Set2')
    axes[i].set_title(f'Accident Counts by {column}', fontsize=16)
    axes[i].set_xlabel(column, fontsize=14)
    axes[i].set_ylabel('Accident Count', fontsize=14)
    axes[i].tick_params(axis='x', labelrotation=45)
    
    # Set the x-axis labels according to the provided labels dictionary
    if column in labels:
        axes[i].set_xticklabels([labels[column][x] for x in sorted(labels[column].keys())])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Accidents_ml = Accidents.drop('Accident_Severity' ,axis=1)
Accidents_ml = Accidents_ml[['Did_Police_Officer_Attend_Scene_of_Accident','Day_of_Week' , 'Weather_Conditions' , 'Road_Surface_Conditions'
                          , 'Light_Conditions','Speed_limit']]

X_train, X_test, y_train, y_test = train_test_split(Accidents_ml.values, 
                                              Accidents['Accident_Severity'].values,test_size=0.20, random_state=99)
```

#### Model-1: Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encode categorical features
encoder = OneHotEncoder(sparse=False)
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)

# Train the Random Forest classifier
random_forest = RandomForestClassifier(n_estimators=200)
random_forest.fit(X_train_encoded, y_train)

# Make predictions and evaluate the model
Y_pred = random_forest.predict(X_test_encoded)
acc_random_forest = round(random_forest.score(X_test_encoded, y_test) * 100, 2)

sk_report = classification_report(
    digits=6,
    y_true=y_test, 
    y_pred=Y_pred)
print("Accuracy" , acc_random_forest)
print(sk_report)
pd.crosstab(y_test, Y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
```

#### Model-2: Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder

# One-hot encode categorical features
encoder = OneHotEncoder(sparse=False)
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)

# Initialize and fit the Logistic Regression model
lr = LogisticRegression()
lr.fit(X_train_encoded, y_train)

# Predict on the test set
y_pred = lr.predict(X_test_encoded)

# Print accuracy and classification report
sk_report = classification_report(
    digits=6,
    y_true=y_test, 
    y_pred=y_pred)
print("Accuracy", round(accuracy_score(y_pred, y_test)*100,2))
print(sk_report)

# Print confusion matrix
pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
```

#### Model-3: Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

# Fit and transform the categorical features in the training set
X_train_encoded = encoder.fit_transform(X_train)

# Transform the categorical features in the test set
X_test_encoded = encoder.transform(X_test)

# Initialize and fit the Decision Tree classifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train_encoded, y_train)

# Predict on the test set
Y_pred = decision_tree.predict(X_test_encoded)

# Calculate accuracy
acc_decision_tree1 = round(decision_tree.score(X_test_encoded, y_test) * 100, 2)

# Print accuracy and classification report
sk_report = classification_report(
    digits=6,
    y_true=y_test, 
    y_pred=Y_pred)
print("Accuracy", acc_decision_tree1)
print(sk_report)

# Print confusion matrix
pd.crosstab(y_test, Y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
```
