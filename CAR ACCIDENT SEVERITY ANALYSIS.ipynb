```python
from sklearn.model_selection import train_test_split
```

```python
Accidents_ml = Accidents.drop('Accident_Severity' ,axis=1)
Accidents_ml = Accidents_ml[['Did_Police_Officer_Attend_Scene_of_Accident','Day_of_Week' , 'Weather_Conditions' , 'Road_Surface_Conditions'
                          , 'Light_Conditions','Speed_limit']]

X_train, X_test, y_train, y_test = train_test_split(Accidents_ml.values, 
                                              Accidents['Accident_Severity'].values,test_size=0.20, random_state=99)
```

#### Model-1: Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encode categorical features
encoder = OneHotEncoder()
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)
```

```python
# Train the Random Forest classifier
random_forest = RandomForestClassifier(n_estimators=200)
random_forest.fit(X_train_encoded, y_train)

# Make predictions and evaluate the model
Y_pred = random_forest.predict(X_test_encoded)
acc_random_forest = round(random_forest.score(X_test_encoded, y_test) * 100, 2)

sk_report = classification_report(
    digits=6,
    y_true=y_test, 
    y_pred=Y_pred)
print("Accuracy" , acc_random_forest)
print(sk_report)
pd.crosstab(y_test, Y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
```

```python
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, Y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", 
            xticklabels=random_forest.classes_,
            yticklabels=random_forest.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")
plt.show()
```

#### Model-2: Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')

# Fit and transform the categorical features in the training set
X_train_encoded = encoder.fit_transform(X_train)

# Transform the categorical features in the test set
X_test_encoded = encoder.transform(X_test)

# Initialize and fit the Decision Tree classifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train_encoded, y_train)

# Predict on the test set
Y_pred = decision_tree.predict(X_test_encoded)

# Calculate accuracy
acc_decision_tree1 = round(decision_tree.score(X_test_encoded, y_test) * 100, 2)

# Print accuracy and classification report
sk_report = classification_report(
    digits=6,
    y_true=y_test, 
    y_pred=Y_pred)
print("Accuracy", acc_decision_tree1)
print(sk_report)

# Print confusion matrix
pd.crosstab(y_test, Y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
```

```python
# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, Y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", 
            xticklabels=decision_tree.classes_,
            yticklabels=decision_tree.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")
plt.show()
```

#### Model-3: Logistic Regression 

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import pandas as pd

# Assuming Accidents is your DataFrame containing the dataset

# One-hot encode categorical features
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(Accidents[['Did_Police_Officer_Attend_Scene_of_Accident', 'Day_of_Week', 'Weather_Conditions', 'Road_Surface_Conditions', 'Light_Conditions']])

# Combine one-hot encoded features with numerical features
X_numerical = Accidents[['Speed_limit']].values
X_combined = pd.concat([pd.DataFrame(X_encoded.toarray()), pd.DataFrame(X_numerical)], axis=1)

# Split the combined data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, Accidents['Accident_Severity'], test_size=0.20, random_state=99)

# Create a pipeline for preprocessing (scaling) and logistic regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Standardize features by removing the mean and scaling to unit variance
    ('classifier', LogisticRegression(random_state=42))  # Logistic Regression classifier
])

# Fit the pipeline (including preprocessing and model) to the training data
pipeline.fit(X_train, y_train)

# Predict on the testing data
y_pred = pipeline.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

## Model Tuning

* Using RandomForestClassifier

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Reduce the size of the dataset
X_train_subset = X_train_encoded[:1000]  # Using the first 1000 samples for training
y_train_subset = y_train[:1000]
X_test_subset = X_test_encoded[:500]  # Using the first 500 samples for testing
y_test_subset = y_test[:500]

# Define a smaller parameter grid to search through
param_dist = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Create a RandomForestClassifier instance
rf = RandomForestClassifier(random_state=99)

# Instantiate RandomizedSearchCV with fewer iterations and folds
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, n_iter=5, cv=3, random_state=99, n_jobs=-1, scoring='accuracy')

# Fit the RandomizedSearchCV instance to the data
random_search.fit(X_train_subset, y_train_subset)

# Print the best parameters found
print("Best Parameters:", random_search.best_params_)

# Get the best model found
best_rf = random_search.best_estimator_

# Evaluate the best model on the test set
test_accuracy = best_rf.score(X_test_subset, y_test_subset)
print("Test Accuracy:", test_accuracy)
```

* Validation Curve

```python
from sklearn.model_selection import validation_curve

# Assuming X_train, y_train are your training data
# Replace these with your actual data if different

# Define the range of hyperparameter values to test
param_range = np.arange(1, 21)

# Create a decision tree classifier
classifier = DecisionTreeClassifier(random_state=42)

# Calculate training and validation scores using validation_curve
train_scores, valid_scores = validation_curve(
    estimator=classifier, 
    X=X_train, 
    y=y_train, 
    param_name="max_depth",  # Hyperparameter to vary
    param_range=param_range,  # Range of hyperparameter values
    cv=5,  # Number of folds in cross-validation
    scoring="accuracy"  # Evaluation metric
)

# Calculate mean and standard deviation of training and validation scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
valid_mean = np.mean(valid_scores, axis=1)
valid_std = np.std(valid_scores, axis=1)

# Plot validation curves
plt.figure(figsize=(10, 6))
plt.plot(param_range, train_mean, label="Training score", color="blue")
plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, alpha=0.2, color="blue")
plt.plot(param_range, valid_mean, label="Cross-validation score", color="red")
plt.fill_between(param_range, valid_mean - valid_std, valid_mean + valid_std, alpha=0.2, color="red")
plt.title("Validation Curve for Decision Tree")
plt.xlabel("Max Depth")
plt.ylabel("Score")
plt.xticks(param_range)
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

```python
# Define the parameter range for which to plot the validation curve
param_range = [100, 200, None, 10, 2, 5, 1, 2]

# Create the validation curve for parameter 'n_estimators'
train_scores, test_scores = validation_curve(
    RandomForestClassifier(random_state=99), 
    X_train_subset, 
    y_train_subset, 
    param_name="n_estimators", 
    param_range=param_range, 
    cv=3, 
    scoring="accuracy", 
    n_jobs=-1
)

# Calculate mean and standard deviation for training set scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)

# Calculate mean and standard deviation for test set scores
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot the validation curve
plt.figure(figsize=(10, 6))
plt.plot(param_range, train_mean, color='blue', marker='o', markersize=5, label='Training Accuracy')
plt.fill_between(param_range, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
plt.plot(param_range, test_mean, color='green', linestyle='--', marker='s', markersize=5, label='Validation Accuracy')
plt.fill_between(param_range, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
plt.title('Validation Curve for RandomForestClassifier')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.xticks(param_range)
plt.grid()
plt.show()
```
